```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 准确率数据示例 (模型 vs 数据集)
data = {
    'Model A': [0.92, 0.85, 0.88],  # 准确率分别为训练集、验证集、测试集
    'Model B': [0.89, 0.84, 0.87],
    'Model C': [0.93, 0.86, 0.90],
    'Model D': [0.91, 0.83, 0.89]
}
```

```python
# 创建 DataFrame
df = pd.DataFrame(data, index=['10-fold_imbalance', '10-fold_balance', 'Stratified sampling'])

# 设置图形尺寸
plt.figure(figsize=(8, 6))

# 使用 seaborn 的 heatmap 函数创建热力图
sns.heatmap(df, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=.5)

# 设置标题
plt.title("Model AUC Comparison Heatmap in the HMDD v4.0 dataset")

# 显示图形
plt.show()
```

```python
# 准确率数据示例（模型在多个数据集上的表现）
data = {
    'l_r': [0.9213, 0.9586, 0.9446, 0.9035]  # 准确率分别对应训练集、验证集、测试集
}

# 数据集名称
datasets = ['1', '2', '3', '4']

# 设置图形尺寸和背景颜色
fig, ax = plt.subplots(figsize=(5, 4), facecolor='white')  # 整体背景为白色

# 设置图表区域背景颜色为白色
ax.set_facecolor('white')

# 为每个模型绘制一条折线
for model, accuracies in data.items():
    ax.plot(datasets, accuracies, marker='o', color='red', label=model, markerfacecolor='red', markeredgecolor='red')
     # 为每个点添加标签
    for i, acc in enumerate(accuracies):
        ax.annotate(f'{acc:.3f}', (datasets[i], acc), textcoords="offset points", xytext=(0,5), ha='center')

# 添加标题和标签
ax.set_title("b.AUC values for different Conv layers", fontsize=12)
ax.set_xlabel("The number of Conv layers", fontsize=10)
ax.set_ylabel("AUC", fontsize=10)
ax.set_ylim(0.7, 1.0)  # 设置 Y 轴范围为 0.8 到 1.0 以突出准确率的差异

# 设置网格线样式为浅紫色虚线，增加类似的网格背景效果
ax.grid(True, color='violet', linestyle='--', linewidth=1.5, alpha=0.3)  # #9370DB 是浅紫色

# 添加顶部的虚线，y=1.0 对应 Y 轴的最大值
ax.axhline(y=1.0, color='violet', linestyle='--', linewidth=3.0)

# 显示图例
# 设置图例的小字体、标记大小和间距
# ax.legend(fontsize='small', markerscale=0.8, labelspacing=0.3, borderpad=0.5)
# plt.savefig('csfx/demo.png', dpi=800, facecolor=fig.get_facecolor())
plt.savefig('csfx/Conv.png')

# 显示图形
plt.show()
```

```python
s = score
```

```python
s
```

```python
for index, row in yz.iterrows():
    miRNA = row[0]
    disease = row[1]
    if miRNA in s.index and disease in s.columns:
        s.at[miRNA, disease] = -1
```

```python
s
```

```python
# 将矩阵展开为长格式，保留miRNA名称和疾病名称
stacked_df = s.stack()
```

```python
# 对分数进行降序排序
sorted_df = stacked_df.sort_values(ascending=False)
```

```python


# 设置调色板为pastel
sns.set_palette("pastel")

plt.figure(figsize=(10, 6))
sns.boxplot(
    x='Epoch', 
    y='Score', 
    hue='Label', 
    data=df, 
    palette="pastel",  # 使用pastel调色板
    flierprops=dict(marker='o', markersize=6, linestyle='none', alpha=0.7)
)

# 设置箱图的透明度
for patch in plt.gca().artists:
    patch.set_alpha(0.5)  # 设置透明度为50%

plt.title('Prediction Score Distribution Across Epochs')
plt.ylabel('Prediction Score')
plt.xlabel('Epoch')

# 保存图像
plt.savefig('train_10_balance_pastel.png')
plt.show()
```
